```python
import random
```

```python
for quiznum in range(2):
    #创建题卡和答案文件
    quiz=open('quiz%s.txt'%(quiznum+1),'w')
    answer=open ('quiz_answer%s.txt'%(quiznum+1),'w')
    #设置题头
    quiz.write((' '*20)+'测试%s'%(quiznum+1))
    quiz.write('姓名:\n\n用时:\n\n分数:\n\n')
    #设置试题  
    for quesnum in range(30):
        #题目一 学前/小学一年级 100以内加减法：
        list_num=[]
        a=random.randint(1,100)
        t=False
        while t==False:
            b=random.randint(1,100)
            if [a,b] not in list_num:
                list_num.append([a,b])
                t=True
            else:
                t=False
        ope_d=random.random()
        if ope_d<=0.5:
            c=a+b
            ope='+'
        else:
            if b>a:
                d=b
                b=a
                a=d
            c=a-b
            ope='-'
        quiz.write('%s. %s %s %s =\n\n\n'%(quesnum+1,a,ope,b))
        answer.write('%s. %s\n'%(quesnum+1,c))
    quiz.close
    answer.close
```

```python
for quiznum in range(100):
    #创建题卡和答案文件
    quiz=open('quiz%s.txt'%(quiznum+1),'w')
    answer=open ('quiz_answer%s.txt'%(quiznum+1),'w')
    #设置题头
    quiz.write('姓名:\n\n用时:\n\n分数:\n\n')
    quiz.write((' '*20)+'测试%s'%(quiznum+1))
    #设置试题  
    for quesnum in range(20):
        #题目二 小学二年级 10以内乘除法：
        list_num=[]
        a=random.randint(1,10)
        t=False
        while t==False:
            b=random.randint(1,10)
            if [a,b] not in list_num:
                list_num.append([a,b])
                t=True
            else:
                t=False
        c=a*b
        ope_d=random.random()
        if ope_d<=0.5:
            ope='x'
            quiz.write('%s. %s %s %s =\n\n\n'%(quesnum+1,a,ope,b))
            answer.write('%s. %s\n'%(quesnum+1,c))
        else:
            ope='➗'
            quiz.write('%s. %s %s %s =\n\n\n'%(quesnum+1,c,ope,a))
            answer.write('%s. %s\n'%(quesnum+1,b))
    quiz.close
    answer.close
```

```python
for quiznum in range(100):
    #创建题卡和答案文件
    quiz=open('quiz%s.txt'%(quiznum+1),'w')
    answer=open ('quiz_answer%s.txt'%(quiznum+1),'w')
    #设置题头
    quiz.write('姓名:\n\n用时:\n\n分数:\n\n')
    quiz.write((' '*20)+'测试%s'%(quiznum+1))
    #设置试题  
    for quesnum in range(20):
        #题目三 小学三年级 100以内乘法列竖式计算：
        list_num=[]
        a=random.randint(1,100)
        t=False
        while t==False:
            b=random.randint(1,100)
            if [a,b] not in list_num:
                list_num.append([a,b])
                t=True
            else:
                t=False
        c=a*b
        line1=str(quesnum+1)+". "+str(a)
        line2=(len(line1)-len(str(b))-2)*" "+"x"+" "+str(b)
        quiz.write('%s\n'%(line1))
        quiz.write('%s\n'%(line2))
        quiz.write('——————\n')
        quiz.write('\n\n\n\n')
        answer.write('%s. %s\n'%(quesnum+1,c))
    quiz.close
    answer.close    
```
